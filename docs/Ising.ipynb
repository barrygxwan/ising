##Perform Analysis

```python
T = list(np.linspace(1,5,1200))
lat = ising.Lattice(10,10,0,T,nghbrs=4,size=5)
```

```python
lat.animate(750,display=False)
```

```python
data = lat.analyze()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(T,data[0],marker='.')
plt.xlabel(r'$T$',fontsize=15)
plt.ylabel(r'$M$',fontsize=15)
plt.xlim(T[0],T[-1])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(T,data[1])
plt.xlabel(r'$T$',fontsize=15)
plt.ylabel(r'$U_{avg}$',fontsize=15)
plt.xlim(T[0],T[-1])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(T,data[2])
plt.xlabel(r'$T$')
plt.ylabel(r'$C_v$')
plt.xlim(T[0],T[-1])
plt.show()
```
